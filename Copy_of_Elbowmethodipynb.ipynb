```python
# First, import all necessary libraries
from sklearn.datasets import load_iris, load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
iris = load_iris()
digits = load_digits()

# Create scalers
scaler = StandardScaler()

# Preprocess IRIS data
iris_scaled = scaler.fit_transform(iris.data)

# Preprocess MNIST data
digits_scaled = scaler.fit_transform(digits.data)

def plot_elbow_method(data, max_k=10, dataset_name=""):
    distortions = []
    K = range(1, max_k + 1)

    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        distortions.append(kmeans.inertia_)

    # Create the elbow plot
    plt.figure(figsize=(10, 6))
    plt.plot(K, distortions, 'bx-')
    plt.xlabel('k (number of clusters)')
    plt.ylabel('Distortion (Inertia)')
    plt.title(f'Elbow Method for Optimal k - {dataset_name}')
    plt.grid(True)
    plt.show()

    return distortions

# Analyze IRIS dataset
print("Analyzing IRIS dataset...")
iris_distortions = plot_elbow_method(iris_scaled, max_k=10, dataset_name="IRIS Dataset")

# Analyze MNIST dataset
print("\nAnalyzing MNIST dataset...")
digits_distortions = plot_elbow_method(digits_scaled, max_k=15, dataset_name="MNIST Digits Dataset")
```
